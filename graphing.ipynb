```python
import matplotlib.pyplot as plt
import json
import numpy as np
plt.close("all")
```

```python
# Parsing JSON
json_data = {"The Learning Partnership": {"detail": {"Mission": "To ensure that all students in Canada have access to the resources and opportunities they need to succeed in school and life.", "PublicPerception": 9, "Impact": "The Learning Partnership has helped over 1 million students in Canada access the resources and opportunities they need to succeed in school and life.", "FundraisingTechnique": "The Learning Partnership raises funds through a variety of channels, including individual donations, corporate sponsorships, and government grants.", "FinancialTransparency": 8}, "spending": {"Program Expenses": 0.75, "Fundraising Expenses": 0.1, "Administrative Expenses": 0.1, "Marketing Expenses": 0.03, "Other Expenses": 0.02}}, "United Way of Alberta": {"detail": {"Mission": "To improve lives and build community by mobilizing the caring power of Albertans.", "PublicPerception": 8, "Impact": "United Way of Alberta invests in programs and services that help Albertans in need, including those related to education.", "FundraisingTechnique": "United Way of Alberta raises funds through a variety of channels, including individual donations, corporate sponsorships, and government grants.", "FinancialTransparency": 7}, "spending": {"Program Expenses": 0.7, "Fundraising Expenses": 0.1, "Administrative Expenses": 0.1, "Other Expenses": 0.05, "Reserves": 0.05}}, "Edmonton Public Schools Foundation": {"detail": {"Mission": "To support and enhance public education in Edmonton.", "PublicPerception": 7, "Impact": "Edmonton Public Schools Foundation has invested over $100 million in public education in Edmonton.", "FundraisingTechnique": "Edmonton Public Schools Foundation raises funds through a variety of channels, including individual donations, corporate sponsorships, and government grants.", "FinancialTransparency": 6}, "spending": {"Program Expenses": 0.8, "Fundraising Expenses": 0.1, "Administrative Expenses": 0.05, "Other Expenses": 0.02, "Reserves": 0.03}}}
```

```python
for agency in json_data:
    # Access spending data from the JSON
    data = json_data[agency]["spending"]

    # Calculate total spending
    total_spending = sum(data.values())

    # Calculate spending percentages
    spending_percentages = {category: (value / total_spending) * 100 for category, value in data.items()}

    # Prepare pie chart data
    pie_chart_data = list(spending_percentages.values())
    pie_chart_labels = list(spending_percentages.keys())

    # Create pie chart
    plt.figure(figsize=(3, 3))
    plt.pie(pie_chart_data, labels=pie_chart_labels, autopct="%1.1f%%")
    plt.title( agency + " Spending Percentages")
    plt.show()
```

```python
agencies = list(json_data.keys())
print(agencies)
```

```python
transparency = np.array([])
perception = np.array([])
```

```python
for i in json_data:
    transparency = np.append(transparency, json_data[i]["details"]["transparency"])
    perception = np.append(perception, json_data[i]["details"]["publicPerception"])
```

```python
weight_counts = {
    "Transparency": transparency,
    "Perception": perception
}
print(transparency, perception)
```

```python
width = 0.5

fig, ax = plt.subplots()
bottom = np.zeros(6)

for boolean, weight_count in weight_counts.items():
    p = ax.bar(agencies, weight_count, width, label=boolean, bottom=bottom)
    bottom += weight_count

ax.set_title("Number of penguins with above average body mass")
ax.legend(loc="upper right")

plt.show()
```
